### 顔の検出

```python
import cv2
from google.colab.patches import cv2_imshow

# Haar-likeカスケード分類機の読み込み
face_cascade = cv2.CascadeClassifier(
    '/content/haarcascade_frontalface_default.xml'
)

# イメージファイルの読み込み
img = cv2.imread('lena_std.tif')
# 顔を検知
face = face_cascade.detectMultiScale(img)

# 検出した顔を矩形で囲む
for (x, y, w, h) in face:
    cv2.rectangle(
        img,
        pt1=(x, y),
        pt2=(x+w, y+h),
        color=(0, 255, 0),
        thickness=2
    )

# 処理後の画像をtempフォルダーに保存
cv2.imwrite('/content/Face_detection.jpg', img)

# 処理後の画像を表示
cv2_imshow(img)

# 何かキーを押したら終了
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 瞳の検出

```python
import cv2
from google.colab.patches import cv2_imshow

# Haar-likeカスケード分類機の読み込み
eye_cascade = cv2.CascadeClassifier(
    '/content/haarcascade_eye.xml'
)

# イメージファイルの読み込み
img = cv2.imread('lena_std.tif')
# 顔の中から瞳を検知
eyes = eye_cascade.detectMultiScale(img)

# 検出した瞳を矩形で囲む
for (x, y, w, h) in eyes:
    cv2.rectangle(
        img,
        pt1=(x, y),
        pt2=(x+w, y+h),
        color=(0, 255, 0),
        thickness=2
    )

# 処理後の画像をtempフォルダーに保存
cv2.imwrite('/content/eyes_detection.jpg', img)

# 処理後の画像を表示
cv2_imshow(img)

# 何かキーを押したら終了
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 検出した部分を切り取って保存する

```python
import cv2
from google.colab.patches import cv2_imshow

# Haar-likeカスケード分類機の読み込み
face_cascade = cv2.CascadeClassifier(
    '/content/haarcascade_frontalface_default.xml'
)

# イメージファイルの読み込み
img = cv2.imread('lena_std.tif')
# 顔を検知
face = face_cascade.detectMultiScale(img)

# 検出した顔を矩形で囲む
for (x, y, w, h) in face:
    trim = img[
        y:y + h,    # y軸の範囲
        x:x + w,    # x軸の範囲
    ]

# 処理後の画像をtempフォルダーに保存
cv2.imwrite('/content/face_01.jpg', trim)

# 処理後の画像を表示
cv2_imshow(trim)

# 何かキーを押したら終了
cv2.waitKey(0)
cv2.destroyAllWindows()
```
